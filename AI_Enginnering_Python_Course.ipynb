```python
# While loops:  A while loop runs until a condition stops being true. We use it less often, but we should know it.

countdown = 10
while countdown > 0:
    print(countdown)
    countdown = countdown - 1 # this is the key part of this.  It tell us what to do with the countdown variable you created outside the loop
print("Go!")

# Every good while loop needs three things, or it will break:
# The Starting Point: count = 3 (Setting up a variable before the loop starts).
# The Condition: while count > 0: (The rule that keeps the loop alive).
# The "Change": count = count - 1 (Changing the variable so the condition eventually becomes False).

#The Golden Rule of while Loops: If you forget to change the variable inside the loop (Step 3),
# the condition will stay True forever. This creates an Infinite Loop, and your program will run endlessly until your computer crashes or you force-close it!

```

```python
# Another Example

def above_average(numbers):
    avg = sum(numbers) / len(numbers)
    result = []
    for n in numbers:
        if n > avg:
            result.append(n)
    return result

scores = [3, 7, 2, 9, 4]
print(f"Average is {sum(scores)/len(scores)}")  # this is just from the scores list
print(f"Above average: {above_average(scores)}")  # here the function is running
```

# Week 8: Data Structures
Lists and dictionaries (Python's two workhorse structures)

A list is an ordered collection of items you reach into by position

A dictionary (dict) Unordered collection of Key-Value pairs.

```python
# List

shopping_list = ["apples", "bananas", "cherries"]

# Dict
room_numbers = {"Alice": 101, "Bob": 102, "Charlie": 103}


# Getting information from list and dictionaries
#List (By Position): To get "bananas" from the list above, you have to ask for its position (index).

print(shopping_list[1])

#To find Bob's room, you don't care where he is in the dictionary; you just look up his name.

print(room_numbers["Bob"])

#List: You can have the same item multiple times. [1, 1, 1, 2] is perfectly fine.
#Dictionary: Keys must be unique. If you try to add a duplicate key, it will just overwrite the old value.
```

💡 Use a List if:

You need to maintain a specific order (e.g., a high score leaderboard, a timeline of events).

Your data is just a collection of single items (e.g., a list of usernames).

💡 Use a Dictionary if:

You need to link data together (e.g., mapping a product ID to its price).

You want to look things up incredibly fast without searching through the whole collection item by item.

```python
# List

teams = ["Texas", "Oklahoma", "Alabama"]

print(teams[0])          # first item — remember, counts from 0
print(teams[-1])         # last item — negative counts from the end
teams.append("Georgia")  # add to the end
print(teams)
print(len(teams))
```

```python
# Dictionaries

conf_sizes = {
    "SEC": 16,
    "Big Ten": 18,
    "ACC": 17,
    "Big 12": 16
}

print(conf_sizes["SEC"])        # look up by key
conf_sizes["Pac-12"] = 2        # add a new pair
print(conf_sizes["Pac-12"])
print(len(conf_sizes))          # how many conferences in dictionary
```

```python
# Looping over a dictionary

for conf, size in conf_sizes.items():    # note we are asking for both conf and size.  Also note conf_size.items is the call.
    print(f"{conf} has {size} teams")


for name, number in room_numbers.items():    # note we are asking for both conf and size.  Also note conf_size.items is the call.
    print(f"{name} is in Room #{number}")
```

```python
# Something to watch out for with Python that would not be an issue with R

a = [1, 2, 3]  # create a list of 3 numbers called a
print(a)

b = a    # assign the values in the a variable to the variable B
print(b)


b.append(4)  # You want to add 4 to the list called B


print("a is now:", a)
print("b is now:", b)

# You changed b, but a changed too. Why: b = a does not make a copy — it
# makes b a second name for the same list. They're one object with two labels.
# Lists, dicts, and DataFrames are mutable (changeable in place), and assignment
# shares them rather than copying. R lulls you into expecting a copy; Python doesn't give you one.

# The fix — make a real copy when you want independence:

a = [1, 2, 3]
b = a.copy()
b.append(4)
print("a is now:", a)   # [1, 2, 3]  ← just the old a list
print("b is now:", b)   # [1, 2, 3, 4]
```

```python
import pandas as pd
import numpy as np

# Set seed for reproducibility
np.random.seed(42)

# Number of quarterback prospects
n_qbs = 50

# Generate prospect names
prospects = [f"QB_{i+1}" for i in range(n_qbs)]

# Generate realistic QB 40-yard dash times
# Mean = 4.78 sec, SD = 0.18 sec
forty_times = np.random.normal(loc=4.78, scale=0.18, size=n_qbs)

# Restrict values to realistic QB range
forty_times = np.clip(forty_times, 4.40, 5.20)

# Generate additional variables
height_inches = np.random.normal(75, 2, n_qbs).round(0)  # ~6'3"
weight_lbs = np.random.normal(225, 12, n_qbs).round(0)
arm_length = np.random.normal(32.5, 1.0, n_qbs).round(1)

# Create DataFrame
qb_data = pd.DataFrame({
    "Prospect": prospects,
    "Height_Inches": height_inches,
    "Weight_Lbs": weight_lbs,
    "Arm_Length": arm_length,
    "Forty_Yard_Dash": forty_times.round(2)
})


print(qb_data.head(10))
qb_data.describe()

# Create a new copy of this dataset
qb_data_51 = qb_data.copy()

new_qb = {
    "Prospect": "QB_51",
    "Height_Inches": 76,
    "Weight_Lbs": 230,
    "Arm_Length": 33.2,
    "Forty_Yard_Dash": 4.68
}

# Append the new row using pd.concat as .append is deprecated
qb_data_51 = pd.concat([qb_data_51, pd.DataFrame([new_qb])], ignore_index=True)


print(qb_data_51)
print(qb_data)
```

```python
# Filtering by Rows


import pandas as pd

df = pd.DataFrame({
    "team": ["Texas", "Alabama", "Oregon", "Kansas", "Duke"],
    "conference": ["SEC", "SEC", "Big Ten", "Big 12", "ACC"],
    "wins": [13, 9, 13, 5, 8],
    "losses": [2, 4, 1, 7, 4]
})

print(df)


print(df["wins"] > 10)          # now we get a true false on whether each row of data has wins over 10


print(df[df["wins"] > 10])      # print rows where wins greater than 10 is true

print(df[(df["wins"] >= 8) & (df["losses"] < 5)]) # you can also set conditions using two conditions.  Here, we
# print rows if wins more than 8 and loses less than 5.  this like the browse of keep if command in STATA

```

```python
# A pandas trap

strong = df[df["wins"] > 10].copy()  # you need to add copy here to make sure that python knows to create a new table for this.  You can then add new variable to that table
print(strong)

strong["note"] = "good team"   # for the rows in strong dataframe, create a new varible called note that says "good team"
print(strong)



# Rule of thumb: when you filter a DataFrame and intend to modify the result, add .copy().
# This single habit eliminates the most common pandas confusion you'll ever hit. You met the
# principle on a list last week; here it is on a table.
```

# Week 12 - Cleaning real data that you create or bring in.

This can be especally important for scrapped data or data you bring in from an API.  Real data work is mostly cleaning

```python
import pandas as pd
df = pd.read_csv('dirty_employee_data.csv')
df

print(df.shape)     # 15 observation, 6 variables
print(df.dtypes)        # <-- the critical diagnostic...the datatype for each variable.
print(df.head())
print(df.isna().sum())  # missing values per column
```

```python
# now you can start cleaning the data


df = pd.read_csv('dirty_employee_data.csv')

# Force a column to numeric; anything unparseable becomes NaN
df["salary"] = pd.to_numeric(df["salary"], errors="coerce")
df["performance_score"] = pd.to_numeric(df["performance_score"], errors="coerce")
df["employee_id"] = pd.to_numeric(df["employee_id"], errors="coerce")

print(df.dtypes)


# make department name consistant
df["department"] = df["department"].str.lower().str.strip()
df

# Drop exact duplicate rows
df = df.drop_duplicates()
df

#Fix the hire dates, which is likely a common problem in data I would download
df['hire_date'] = pd.to_datetime(df['hire_date'], format='mixed')
df

#I can also format the dates in a diffrent way if I wanted
df['hire_date_words'] = df['hire_date'].dt.strftime('%B %d, %Y')  #created a new variable called dataframe words
df
```

```python
# Now I want to get the mean salary per department with counts of the number of
# employees by department and the mean number of days the person has worked at the company.
# I can now ask an LLM to do all this.

# Calculate the number of days worked as of June 5, 2026
target_date = pd.to_datetime('2026-06-05')
df['days_worked'] = (target_date - df['hire_date']).dt.days
df

summary_df = df.groupby('department').agg(
    employee_count=('employee_id', 'count'),
    mean_salary=('salary', 'mean'),
    mean_days_worked=('days_worked', 'mean')
).reset_index()

summary_df['mean_salary'] = summary_df['mean_salary'].round().astype(int)
summary_df['mean_days_worked'] = summary_df['mean_days_worked'].round().astype(int)

print(summary_df)

# Format 'mean_salary' with a comma and 0 decimal places
styled_df = summary_df.style.format({
    'mean_salary': '{:,.0f}',
    'mean_days_worked': '{:,.0f}'
})

styled_df
```

```python
# Another Example

%%writefile balance.py

def win_pct(wins, losses):
    return wins / (wins + losses)

def classify(pct):
    if pct > 0.7:
        return "strong"
    elif pct > 0.5:
        return "winning"
    else:
        return "losing"

def main():
    teams = {"Texas": (13, 2), "Kansas": (5, 7), "Duke": (8, 5), "SMU": (12, 1)}
    for team, (wins, losses) in teams.items():
        pct = win_pct(wins, losses)
        print(f"{team}: {pct:.3f} ({classify(pct)})")

main()
```

```python
## A practice script related to my work in higher education


%%writefile faculty_audit.py

def check_sabbatical(years_service, last_sabbatical_years_ago):
    """Helper Function 1: Determines if a faculty member is eligible for sabbatical."""
    # Eligible if they've served at least 6 years AND haven't had a sabbatical in the last 6 years
    if years_service >= 6 and last_sabbatical_years_ago >= 6:
        return "ELIGIBLE"
    else:
        return "Not Eligible"

def evaluate_productivity(publications, grants_secured):
    """Helper Function 2: Categorizes research productivity for annual reviews."""
    if publications >= 3 and grants_secured >= 1:
        return "High"
    elif publications >= 1 or grants_secured >= 1:
        return "Satisfactory"
    else:
        return "Needs Review"

def main():
    """Main Function: Processes the faculty roster and prints the Dean's report."""
    # Simulated data database pull: { "Faculty Name": (Years Service, Years Since Last Sabbatical, Pubs, Grants) }
    faculty_roster = {
        "Dr. Aris Vance": (12, 7, 4, 1),
        "Dr. Elena Rostova": (4, 4, 2, 0),
        "Prof. Marcus Cole": (8, 2, 0, 0),
        "Dr. Aisha Rahman": (15, 8, 0, 2)
    }

    print("=====================================================================")
    print("          FACULTY AFFAIRS AFFAIRS: ANNUAL AUDIT REPORT               ")
    print("=====================================================================\n")

    # Loop through each faculty member and process their data
    for name, data in faculty_roster.items():
        years_service, last_sab, pubs, grants = data

        # Call Helper Function 1
        sabbatical_status = check_sabbatical(years_service, last_sab)

        # Call Helper Function 2
        research_status = evaluate_productivity(pubs, grants)

        # Print the structured output for the Dean's review
        print(f"Faculty: {name}")
        print(f"  - Sabbatical Status:      {sabbatical_status}")
        print(f"  - Research Productivity:  {research_status}")
        print("-" * 50)

# The Trigger: This executes the main function when the script runs
if __name__ == "__main__":
    main()
```
